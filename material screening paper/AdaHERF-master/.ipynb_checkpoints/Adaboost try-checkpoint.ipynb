###Adaboost 

Load required libraries

```python
import pandas as pd
import numpy as np
import scipy as sp

from sklearn.cross_validation import StratifiedKFold
from sklearn.decomposition import RandomizedPCA
from sklearn.decomposition import PCA
# from AdaHERF import AdaHERF 
```

Load data and choose the X and Y

```python
binary_sys = pd.read_pickle('../binary_sys')
binary_sys['type'] = 'binary'
ternary_sys = pd.read_pickle('../tern_data')
ternary_sys['type'] = 'ternary'

data = pd.concat([binary_sys,ternary_sys], ignore_index= True)

features = list(data.columns)
features = [f for f in features if (f.endswith('prop') 
                                    or f.startswith('prop') 
                                    or f.startswith(('avg','max')) 
                                    or f in 'ABC' 
                                    or f == 'formation_energy'
                                    or f == 'type')]
data = data[features]
data = data.fillna(0)
data.reindex()
Y = 'formation_energy'
X = list(set(features) - set(['formation_energy','type','A','B','C']))

```

Test Adaboost: 
* base algorithm = Decision Tree
* Max depth of tree = 6
* Number of estimators = 300
               

```python
# test ADAboost 

%matplotlib inline

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor

from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt

rng = np.random.RandomState(93)
X_train,X_test,Y_train,Y_test = train_test_split(data[X],data[Y],test_size = 0.33,
                                                 random_state = 93)

adaboost = AdaBoostRegressor(DecisionTreeRegressor(max_depth=6),
                          n_estimators=300, random_state=rng)


fit = adaboost.fit(X_train,Y_train)
predict = adaboost.predict(X_test)


```

Plot Predicted Vs measured

```python
import seaborn as sns
sns.set_style("darkgrid")

fig,ax = plt.subplots()
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
ax.scatter(Y_test, predict,)
ax.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
fig.show()
```
